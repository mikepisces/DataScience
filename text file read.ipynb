```python
##Read the tst file
import matplotlib.pyplot as plt

emp_list = []
with open(r"C:\Users\Khushi\Documents\modi_speech.txt") as fileinput:
    for line in fileinput:
        line = line.lower()
        emp_list.append(line)
        
emp_list        
```

```python
from wordcloud import WordCloud
```

```python
cloud = WordCloud().generate(str(emp_list))
```

```python
plt.figure(figsize = (15,15)) ## just to increase the size of the plot
plt.imshow(cloud)
```
